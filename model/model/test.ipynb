```python
import glob

pill_df = pd.read_csv("../Data/pill_kr_reduce.csv", encoding='utf-8')
pill_df = pill_df[pill_df.ETC_OTC_NAME == '일반의약품']
pill_df.head()
```

```python
filepath_front = sorted(glob.glob('../Data/separate/front/*.png'))
filepath_back = sorted(glob.glob('../Data/separate/back/*.png'))
fail_num = 0
correct_num = 0

for i, row in pill_df.head(500).iterrows():
    img1 = cv.imread(filepath_front[i])
    img2 = cv.imread(filepath_back[i])
    top5 = pill_classification_top5(img1, img2)
    if i == top5[0]['ITEM_INDEX']:
        correct_num += 1
    else:
        fail_num += 1
    print(f'정답 클래스: {i}, 예측 클래스: {top5[0]["ITEM_INDEX"]}')
print(f'total: {fail_num+correct_num}, correct: {correct_num}, accuracy: {correct_num/(fail_num+correct_num)*100}%')
```
